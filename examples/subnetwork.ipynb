```python
import numpy as np
import pandas as pd
import tensorflow as tf
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.utils import shuffle
import scipy.stats as stats

import tensorflow as tf
from xnn.layers import *
from xnn.gamnet import GAMNet

np.set_printoptions(suppress=True, precision = 4)
```

```python
def gen_trun_normal(DataNum, lower, upper, mu= 0, sigma = 0.1):
    X = stats.truncnorm(
        (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
    N = stats.norm(loc=mu, scale=sigma)
    return X.rvs(DataNum)

def doppler(DataNum, noise_level):
    x = np.sort(gen_trun_normal(DataNum, 0, 1, 0.5, 0.3) )
    x = np.linspace(0, 1, DataNum)
    y = np.sin(x*np.pi*2)+ noise_level*np.random.normal(0, 1, DataNum) 
#     y = np.sqrt(x*(1-x))*np.sin((2.1*np.pi)/(x+0.05))+ noise_level*np.random.normal(0, 1, DataNum) 
    return x, y 
```

```python
DataNum = 3000
X, Y = doppler(DataNum, 0.1)
X = X.reshape([-1,1])
Y = Y.reshape([-1,1])
plt.plot(X,Y)
```

```python
tf.random.set_seed(0)
model = GAMNet(input_num = 1, input_dummy_num=0, subnet_arch=[10, 6], task="Regression",
               activation_func=tf.tanh, batch_size=3000, training_epochs=300, lr_bp=0.01,
               beta_threshold=0.05, tuning_epochs=0, l1_subnet=0.001, smooth_lambda=0.0001,
               verbose=True, val_ratio=0.0, early_stop_thres=200)
model.fit(X, Y)
model.visualize("./", "test")
```

```python
plt.plot(X, Y)
plt.plot(X, model.predict(X).numpy())
```
